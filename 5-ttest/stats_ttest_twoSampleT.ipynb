```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
## generate the data

# parameters
n1 = 30   # samples in dataset 1
n2 = 40   # ...and 2
mu1 = 1   # population mean in dataset 1
mu2 = 1.2 # population mean in dataset 2


# generate the data
data1 = mu1 + np.random.randn(n1)
data2 = mu2 + np.random.randn(n2)

# show their histograms
plt.hist(data1,bins='fd',color=[1,0,0,.5],label='Data 1')
plt.hist(data2,bins='fd',color=[0,0,1,.5],label='Data 2')
plt.xlabel('Data value')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
## now for the t-test

t,p = stats.ttest_ind(data1,data2,equal_var=True) # The 'ind' in ttest_ind is for independent/unpaired samples

df = n1+n2-2
print('t(%g) = %g, p=%g'%(df,t,p))
```

```python
## a 2D space of t values

# ranges for t-value parameters
meandiffs = np.linspace(-3,3,80)
pooledvar = np.linspace(.5,4,100)

# group sample size
n1 = 40
n2 = 30

# initialize output matrix
allTvals = np.zeros((len(meandiffs),len(pooledvar)))

# loop over the parameters...
for meani in range(len(meandiffs)):
    for vari in range(len(pooledvar)):
        
        # t-value denominator
        df = n1 + n2 - 2
        s  = np.sqrt(( (n1-1)*pooledvar[vari] + (n2-1)*pooledvar[vari]) / df)
        t_den = s * np.sqrt(1/n1 + 1/n2)
        
        # t-value in the matrix
        allTvals[meani,vari] = meandiffs[meani] / t_den

        
plt.imshow(allTvals,vmin=-4,vmax=4,extent=[pooledvar[0],pooledvar[-1],meandiffs[0],meandiffs[-1]],aspect='auto')
plt.xlabel('Variance')
plt.ylabel('Mean differences')
plt.colorbar()
plt.title('t-values as a function of difference and variance')
plt.show()
```

```python
import seaborn as sns
 
mean1 = 1
mean2 = 1.2
num_exps = 100
pval_thresh = .05
std = .5
 
results = []
 
for samp_size in range(10, 201):
    ss_results = []
    for rep in range(num_exps):
        data1 = np.random.randn(samp_size) * std + mean1
        data2 = np.random.randn(samp_size) * std + mean2
        t, p = stats.ttest_ind(data2, data1, equal_var=False)
        ss_results.append(t)
        
    results.append(ss_results)
        
means = [np.mean(x) for x in results]
 
# t score thresholds for one tailed test .05% use ppf to get 95% threshold
t_thresh = []
for ss in range(10, 201):
    df = 2 * ss - 2
    t_thresh.append(stats.t.ppf(1 - pval_thresh, df))
 
sns.set_style('white')
plt.figure(figsize=(10,6))
plt.plot(range(10, 201), results, color=[0.6, 0.6, 0.6, 0.2])
plt.plot(range(10, 201), means, 'k-', label="Mean of 100 T-scores taken for each sample size")
plt.plot(range(10, 201), t_thresh, 'r-', label="T-score Threshold .05")
plt.ylim(-4, 8)
 
plt.title("Effect of Sample Size on T-scores\n(two sample t-tests with means 1 and 1.2)", fontsize=16)
plt.xlabel("Sample Size (Equal Groups)", fontsize=12)
plt.ylabel("T-test value", fontsize=12)
plt.legend()
plt.show()
```
